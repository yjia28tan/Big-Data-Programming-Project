```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
#to ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Load datasets
customers = pd.read_csv("olist_customers.csv")
geolocation = pd.read_csv('olist_geolocation.csv')
order_items = pd.read_csv('olist_order_items.csv')
payments = pd.read_csv('olist_payments.csv')
reviews = pd.read_csv('olist_reviews.csv')
orders = pd.read_csv('olist_orders.csv')
products = pd.read_csv('olist_products.csv')
sellers = pd.read_csv('olist_sellers.csv')
```

```python
# Merge datasets
order_customer_geolocation = customers.merge(orders, on='customer_id').merge(geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
order_product_info = order_items.merge(products, on='product_id').merge(order_customer_geolocation, on='order_id')  
order_product_info
```

```python
# Group by customer location and calculate the total quantity of ordered products
location_product_data = order_product_info.groupby(['customer_city', 'customer_state', 'customer_zip_code_prefix', 'geolocation_lat', 'geolocation_lng','order_purchase_timestamp'])['product_id'].count().reset_index()
location_product_data
```

```python
# Find the top 1 product category for each customer city
top_category_per_city = location_product_data.sort_values('product_id', ascending=False).groupby('customer_city').head(1)

# Display the result
top_category_per_city
```

```python
# Assuming you have a DataFrame with the product category information for each customer city
location_product_data['product_id'] = location_product_data.groupby('customer_city')['product_id'].transform(lambda x: x.mode().iloc[0])

# Filter to keep only the rows with the top product category for each city
top_category_per_city = location_product_data.drop_duplicates(subset=['customer_city', 'product_id'])

# Assuming you have a GeoDataFrame with geometries for customer locations
geometry = gpd.points_from_xy(top_category_per_city['geolocation_lng'], top_category_per_city['geolocation_lat'])
gdf = gpd.GeoDataFrame(top_category_per_city, geometry=geometry, crs="EPSG:4326")
# Load a map of the world
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter for South America, to zoom into Brazil
ax = world[world.continent == "South America"].plot(color="white", edgecolor="black", figsize=(15, 15))

# Set up a categorical color palette
palette = sns.color_palette("Set3", n_colors=len(top_category_per_city['product_id'].unique()))

# Scatter plot with color representing product category and size representing product count
sc = plt.scatter(gdf['geolocation_lng'],
                gdf['geolocation_lat'],
                c=gdf['product_id'].astype('product').cat.codes,
                s=gdf['product_id'],
                cmap='Set3', 
                alpha=0.7)

# Create a legend
legend_labels = gdf['product_id'].unique()
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[i], markersize=10, label=label) for i, label in enumerate(legend_labels)]
plt.legend(handles=legend_handles, title='Product')

# Set axis labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Top Product per Customer City')

plt.show()
```

```python
# Convert 'order_purchase_timestamp' to datetime
order_customer_geolocation['order_purchase_timestamp'] = pd.to_datetime(order_customer_geolocation['order_purchase_timestamp'])

# Extract the year from the timestamp
order_customer_geolocation['order_year'] = order_customer_geolocation['order_purchase_timestamp'].dt.year
```

```python
order_customer_geolocation
```

```python
# Filter data for each year
for year in range(2016, 2019):
    # Filter data for the current year
    current_year_data = order_customer_geolocation[order_customer_geolocation['order_year'] == year]

    # Assuming you have a DataFrame named 'location_product_data' with 'geolocation_lat', 'geolocation_lng', and 'product_category_name' columns
    gdf = gpd.GeoDataFrame(current_year_data, geometry=gpd.points_from_xy(current_year_data['geolocation_lng'], current_year_data['geolocation_lat']))

    # Plotting
    ax = world[world.continent == "South America"].plot(color="white", edgecolor="black", figsize=(15, 15))
    gdf.plot(ax=ax, marker='o', color='red', markersize=5, alpha=0.5)
    plt.title(f'Customer Location and Product Categories - {year}')
    plt.show()
    
```

```python
product_category_data = order_product_info.groupby(['customer_city', 'customer_state', 'customer_zip_code_prefix', 'product_category_name'])['product_id'].count().reset_index()
product_category_data.columns = ['customer_city', 'customer_state', 'customer_zip_code_prefix', 'product_category_name', 'product_count']
```

```python
# Create a GeoDataFrame for customer locations
customer_geo_df = gpd.GeoDataFrame(location_product_data, 
                                   geometry=gpd.points_from_xy(location_product_data['geolocation_lng'], location_product_data['geolocation_lat']))

# Merge customer_geo_df with product_category_data
customer_product_geo_df = pd.merge(customer_geo_df, product_category_data, on=['customer_city', 'customer_state', 'customer_zip_code_prefix'], how='left')
```

```python
# Create a scatter plot
fig, ax = plt.subplots(figsize=(10, 8))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# Scatter plot with color representing product category and size representing product count
sc = ax.scatter(customer_product_geo_df['geolocation_lng'], 
                customer_product_geo_df['geolocation_lat'], 
                c=customer_product_geo_df['product_category_name'].astype('category').cat.codes,
                s=customer_product_geo_df['product_count'],
                cmap='viridis',
                alpha=0.7)

# Colorbar
cbar = plt.colorbar(sc, cax=cax)
cbar.set_label('Product Category')

# Set axis labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Customer Location and Product Categories')

plt.show()
```
